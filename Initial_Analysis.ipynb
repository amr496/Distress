## Import Data Based on Query
This should be limited to 1 filename. Can use below with ONLY changing **filename** variable

```python
query = '''SELECT CONCAT(CAST("MP" AS VARCHAR), '_' , "Direction") AS "Measurement_Sections",
"Test_Date",
"Filename",
"Pulse_Num",
"Time",
"Project",
"Pavement",
"SRI",
"MP",
"Direction",
"Vehicle_Speed",
left("Vehicle_Speed",2) AS "Vehicle_Speed_Corr",
"400_Hz" AS "Hz400", 
"500_Hz" AS "Hz500", 
"630_Hz" AS "Hz630", 
"800_Hz" AS "Hz800", 
"1000_Hz" AS "Hz1000", 
"1250_Hz" AS "Hz1250", 
"1600_Hz" AS "Hz1600", 
"2000_Hz" AS "Hz2000", 
"2500_Hz" AS "Hz2500", 
"3150_Hz" AS "Hz3150", 
"4000_Hz" AS "Hz4000", 
"5000_Hz" AS "Hz5000", 
"Awtd", 
"Handwritten_Values",
"Tire",
"Air_Temperature",
ABS("Handwritten_Values"-"Awtd") AS "Absolute_Value_Hand" 


FROM public."OBSIData"
WHERE "Good_Bad" = 'G'
ORDER BY "Measurement_Sections" ASC;'''


df_noise=pd.read_sql_query(query,con=conn)
count_records = len(df_noise.index)
print ("The total record count of this datatset is:", count_records)
```
